# Ejemplo de uso de ecuaciones LaTEX

La función sigmoide $\sigma(x)$ se define según se indica en la fórmula siguiente:

\begin{equation}
\sigma(x)= {1\over {1+e^{-x}}}
\end{equation}

Algunas de las funciones de activación más comunes son:


\begin{array}{cc}
\textbf{Nombre} & \textbf{Función}\\
\hline
Sigmoide & \sigma(x)= {1\over {1+e^{-x}}}\\
ReLU & f(x)= max(0, x)\\
... & ...
\end{array}
