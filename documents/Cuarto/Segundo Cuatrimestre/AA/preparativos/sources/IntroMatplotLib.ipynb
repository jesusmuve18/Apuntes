# **Figuras y gráficas**


Aunque MatplotLib sea una biblioteca específica de generación de gráficas, el paquete más usado de la biblioteca se corresponde con el paquete **PyPlot**.

## El paquete PyPlot

El paquete PyPlot se suele importar desde MatplotLib con un alias:

````python
import matplotlib.pyplot as plt
````

La generación de figuras se realiza de forma simple desde la función **figure**, la cual tiene diversos argumentos opcionales (tamaño, disposición, etc.)

````python
f= plt.figure()

# se suele usar la función show para mostrar las figuras
plt.show()

# O también usando display()
display(f) # En entornos distintos de IPython,



````

Se puede configurar (prácticamente) todo dentro de una figura:

- Tamaño de las diversas fuentes
- Etiquetas de los ejes
- Título de la imagen
- etc., etc., etc.


Un ejemplo:

````python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(start=0, stop= 5, num=11)
y= np.random.rand(11)

plt.figure()
plt.plot(x, y) # Muestra los valores x frente a y
plt.xlabel('Eje X') # Etiqueta del eje X
plt.ylabel('Eje Y') # Etiqueta del eje X
plt.title('Título') # Título de la figura
plt.show()
````

Se pueden agrupar varias gráficas dentro de la misma figura. Por ejemplo:

````python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(start=0, stop= 5, num=11)
y= np.random.rand(11)
z= np.random.randn(11)

plt.figure()
plt.plot(x, y) # Muestra los valores x frente a y
plt.plot(x, z) # Muestra los valores x frente a y
plt.xlabel('Eje X') # Etiqueta del eje X
plt.ylabel('Eje Y') # Etiqueta del eje X
plt.title('Título') # Título de la figura
plt.legend(['Uniforme', 'Normal']) # Legenda de colores
plt.show()
````

Por supuesto, también se puede seleccionar el color a mostrar en cada gráfica, así como el formato en el que se muestran los datos.

Para más información, véase la API en <a href="">https://matplotlib.org/stable/api/index.html</a>

Un ejemplo:

````python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(start=0, stop= 5, num=11)
y= np.random.rand(11)
z= np.random.randn(11)

plt.figure()
plt.plot(x, y, 'r') # Muestra los valores x frente a y en rojo
plt.plot(x, z, 'b*') # Muestra los valores x frente a y en azul con * en lugar de líneas
plt.xlabel('Eje X') # Etiqueta del eje X
plt.ylabel('Eje Y') # Etiqueta del eje X
plt.title('Título') # Título de la figura
plt.legend(['Uniforme', 'Normal']) # Legenda de colores
plt.show()
````

# **Sub-figuras**

Se puede crear una figura general que agrupe varias figura. Suele ser útil cuando se desea mostrar resultados referentes a un mismo experimento.

Por ejemplo

````python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(start=0, stop= 5, num=11)
y= np.random.rand(11)

colores= ('red', 'blue', 'black', 'green')


f= plt.figure()
for i, color in enumerate(colores):
  plt.subplot(2, 2, i+1)
  plt.plot(x, y, color)

plt.show()
````

## Cuestiones de formato

Existen multitud de opciones dentro de MatplotLib para seleccionar el formato adecuado. Se puede hacer (casi) todo lo que deseemos (el problema, a veces, resulta en saber *cómo* hacerlo).

Un ejemplo de formato algo más avanzado:

````python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(start=-10, stop= 10, num=101)
y_exp= np.exp(x)
y_log= np.log(x)

f, ax= plt.subplots() # Figura con subfiguras. Devuelve la imagen f y sus ejes
ax.plot(x, y_exp, 'k--', label= 'y=exp(x)')
ax.plot(x, y_log, 'k:', label= 'y=ln(x)')

# Localización de la leyenda
legend= ax.legend(loc='lower right', shadow= True, fontsize='x-large')

# Ponemos límites al mostrar los ejes en -5, 5
ax.set_ylim( (-5, 5) )

# Ponemos color de fondo a la leyenda
legend.get_frame().set_facecolor('C0')

# Ponemos un grid para mejorar visualización
plt.grid()

plt.show()
````

# **Algunos formatos de gráficos adicionales**

## Gráficos de barras

````python
import numpy as np
import matplotlib.pyplot as plt

# Variable aleatoria
X= ['cara', 'cruz']

# Probabilidades de variable aleatoria
p= [0.25, 0.75]

# Muestreo
muestras= np.random.choice(a= X, size=100, p= p)
n_caras= np.sum(muestras==X[0])
n_cruces= np.sum(muestras==X[1])

plt.bar(X, [n_caras, n_cruces])
plt.show()
````

## Histogramas

````python
import numpy as np
import matplotlib.pyplot as plt

# Variable aleatoria
muestras= np.random.randn(10000)

plt.hist(muestras, bins=10); # OJO: Aquí el ; sirve para que no muestre el texto de la figura
plt.show();
````

## Gráficos de dispersión

````python
import numpy as np
import matplotlib.pyplot as plt

# Variable aleatoria
X= np.linspace(0, 100, 101)
Y= X + 100*(np.random.rand(101)-0.5)

Y_sup= Y>X

# Nube de puntos separando los que quedan por encima de Y=X de los que no
plt.scatter(X, Y, c= Y_sup)

# Línea de guía Y=X
plt.plot(X, X, 'k--')

plt.show()
````

* Implemente una función **def aproxima_pi(pares, pares_dentro)** que, teniendo como entrada las salidas de las dos funciones anteriores, calcule como salida una aproximación del número $\pi$.

Suponiendo que ya están implementadas las funciones anteriores, adapte el psudocódigo siguiente para que se ejecute un total de 100 ejecuciones distintas para aproximar $\pi$ considerando un valor de $n=100$.

```
pares_por_ejecucion= [] # Lista de lista de pares obtenidos en cada ejecución
pares_dentro= [] # Lista de lista de pares dentro del círculo en cada ejecución
aproximacion_pi= [] # Aproximación de pi en cada ejecución
desviacion= [] # Desviación de la aproximación del valor real de pi en cada ejecución
error= [] # Error cometido (medido como valor absoluto de la desviación)
num_experimentos= 30
n= 100

for i in range(num_experimentos):
  pares_actual= generador(n)
  pares_dentro_actual= dentro(pares_actual)
  aproximacion_actual= aproxima_pi(pares_actual, pares_dentro_actual)

  pares_por_ejecucion.extend(pares_actual)
  pares_dentro.extend(pares_dentro_actual)
  aproximacion_pi.append(aproximacion_actual)
  desviacion.append( np.pi - aproximacion_actual )
  error.append( np.abs(np.pi - aproximacion_actual) )
```
