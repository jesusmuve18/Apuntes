# **Tipos de datos básicos**

Los tipos de datos básicos existentes en Python son los siguientes:

* Números enteros (tipo ```int```). Literales: 1, -2, 7, -9, ...
* Números reales (tipo ```float```). Literales: 1.76, -7.98, 0.01, ...
* Números complejos (tipo ```complex```). Literales: 1+0.j, 1.3j, 10-7j, etc.
* Valores booleanos (tipo ```bool```). Literales: True, False **OJO: Python es sensible a mayúsculas**. Los literales booleanos se escriben **True** y **False**, y no como en C. **NO** se escribe **~~true~~** ni tampoco **~~false~~**.
* Cadenas de caracteres (tipo ```str```). Literales: 'Una cadena', "Una cadena", """Cadena larga de varias líneas""", '''Cadena larga de varias líneas''', etc.
* El valor sin tipo **None** (escrito así, literalmente ```None```).

La siguiente celda contiene ejemplos de literales y del uso de la función ```type``` para conocer su tipo.

````python
# Ejemplos de literales numéricos y valor None
print('Literal entero: ', 10, 'de tipo', type(10))
print('Literal real: ', 10.0, 'de tipo', type(10.0))
print('Literal complejo: ', 10+2.j, 'de tipo', type(10+2.j))
print('Literal booleano: ', True, 'de tipo', type(True))
print('Valor None: ', None, 'de tipo', type(None))

# Ejemplos de literales cadena
print('\n\n') # Los caracteres especiales como retorno de carro, tabulador, etc., se escriben igual que en C
print('Literales sobre cadenas\n')
print('Literal cadena (I): ', 'cadena', 'de tipo', type('cadena'))
print('Literal cadena (II): ', "cadena", 'de tipo', type("cadena"))

# Cadenas de caracteres literales de varias líneas
una_cadena_larga= """Esto es
una cadena de caracteres
de tres líneas"""

print('\nLiteral cadena (III): ', una_cadena_larga, 'de tipo', type(una_cadena_larga))

una_cadena_larga= '''Esto es
una cadena de caracteres
de tres líneas'''

print('\nLiteral cadena (IV): ', una_cadena_larga, 'de tipo', type(una_cadena_larga))
````

## Operadores aritméticos

En Python disponemos de todos los operadores aritméticos comunes:

* Suma (```+```). Ejemplo: a+b
* Resta (```-```). Ejemplo: a-b
* Opuesto (```-```). Ejemplo: -a
* División (```/```). Ejemplo: 1/2 (resultado 0.5)
* División entera (```//```). Ejemplo: 1//2 (resultado 0)
* Módulo (```%```). Ejemplo:  1%0.9 (resultado 0.1). **OJO: El módulo no sólo opera sobre enteros, sino que puede operar sobre reales también**.
* Exponenciación (```\*\*``). Ejemplo: 2\*\*3 (resultado 8).

````python
# Ejemplos de uso de operadores aritméticos
suma= 3+4
resta= 3-4
opuesto= -4
division= 3/4
division_entera= 3//4
modulo= 3%2
exponenciacion= 2**3

print('suma vale', suma)
print('resta vale', resta)
print('opuesto vale', opuesto)
print('division vale', division)
print('division_entera vale', division_entera)
print('modulo vale', modulo)
print('exponenciacion vale', exponenciacion)
````

Python incorpora de forma nativa algunos operadores más, como por ejemplo el cálculo del máximo y del mínimo de dos elementos. La siguiente celda muestra un ejemplo de su uso.

````python
nota_examen_teoria_1= 4.4
nota_examen_teoria_2= 7.4
nota_examen_practicas_1= 7.8
nota_examen_practicas_2= 8.7

nota_teoria= max(nota_examen_teoria_1, nota_examen_teoria_2)
nota_practicas= max(nota_examen_practicas_1, nota_examen_practicas_2)

nota_final= min(nota_teoria, nota_practicas)
print('La nota final es', nota_final)
````

## Operadores booleanos

Python dispone de los operadores booleanos clásicos **and**, **or** y **not**. Con ellos se pueden construir expresiones lógicas más complejas como se muestra en el siguiente ejemplo.

````python
aprobar_examen_teoria_1= True
aprobar_examen_teoria_2= True
aprobar_examen_practicas_1= False
aprobar_examen_practicas_2= True

aprobar_practicas= aprobar_examen_practicas_1 or aprobar_examen_practicas_2
aprobar_teoria= aprobar_examen_teoria_1 and aprobar_examen_teoria_2
aprobar_asignatura= aprobar_teoria and aprobar_practicas
necesita_recuperar= not (aprobar_examen_teoria_1 and aprobar_examen_teoria_2 and aprobar_examen_practicas_1 and aprobar_examen_practicas_2)

print('El estudiante ha aprobado las prácticas: ', aprobar_practicas)
print('El estudiante ha aprobado la teoría: ', aprobar_teoria)
print('El estudiante puede recuperar alguna parte: ', necesita_recuperar)
print('El estudiante ha aprobado la asignatura: ', aprobar_asignatura)

matricula_honor= aprobar_asignatura and not necesita_recuperar
print('El estudiante tiene M.H.: ', matricula_honor)

````

## Operadores relacionales

Son los típicos operaciones que nos permiten realizar comparaciones entre valores de tipos de datos comparables. Se corresponden con los ya conocidos *menor*, *mayor*, *menor o igual*, *mayor o igual*, *igual* y *distinto*. Devuelven un valor booleano indicando si se cumple la condición (**True**) o no (**False**)

* Menor (```<```). Ejemplo: 2<3 -> True
* Mayor (```>```). Ejemplo: 2>3 -> False
* Menor o igual (```<=```). Ejemplo: 2<=3 -> True
* Mayor o igual (```>=```). Ejemplo: 2>=3 -> False
* Igual (```==```). Ejemplo: 2==3 -> False
* Distinto (```!=```). Ejemplo: 2!=3 -> True

La celda siguiente contiene ejemplos sobre cómo utilizar estos operadore, también en combinación con operadores lógicos.

````python
calificacion_examen_teoria_1= 5.6
calificacion_examen_teoria_2= 6.5
calificacion_examen_practicas_1= 9.6
calificacion_examen_practicas_2= 6.9

aprobar_teoria= (calificacion_examen_practicas_1 + calificacion_examen_practicas_2)/2 >= 5
aprobar_practicas= (calificacion_examen_practicas_1 + calificacion_examen_practicas_2)/2 >= 5

print('El alumno aprueba teoría: ', aprobar_teoria)
print('El alumno aprueba prácticas: ', aprobar_practicas)

nota_media= calificacion_examen_teoria_1 + calificacion_examen_teoria_2 + calificacion_examen_practicas_1 + calificacion_examen_practicas_2
nota_media/= 4
print('La nota media del alumno es', nota_media)

opta_matricula_honor= nota_media >= 9 and nota_media <= 10
print('El alumno opta a M.H.: ', opta_matricula_honor)

# Otra forma de calcular si un valor se encuentra en un rango
print('El alumno tiene calificación suspenso: ', nota_media < 5)
print('El alumno tiene calificación aprobado: ', 5<= nota_media < 7)
print('El alumno tiene calificación notable: ', 7<= nota_media < 9)
print('El alumno tiene calificación sobresaliente: ', 9<= nota_media <= 10)

````

Además de los operadores relacionales clásicos, Python contempla otros operadores adicionales:

* Operador **is**. Se utiliza para conocer si dos variables distintas apuntan al mismo objeto. Se diferencia de ``==``` en que **is** comprueba si ambas variables apuntan al mismo objeto mientras que ```==``` comprueba si su valor es el mismo. Se puede combinar con el operador lógico ```not``` como **is not** para comprobar la condició  opuesta.
* Operador **in**. Equivale a la comprobación de pertenencia de un elemento a un conjunto. Devuelve **True** si un objeto/valor está incluido en una colección de elementos y **False** en caso contrario. También se puede combinar con el operador lógico ```not``` aunque, al contrario que **is**, la combinación debe escribirse como **not in**



````python
# Un ejemplo de uso de is / is not
x= 1.0
y= None

print('valor de x', x)
print('valor de y', y)

z= x is None or y is None
print('O bien x es None, o bien y es None (o ambos): ', z)

z= x is not None and y is not None
print('Ni x ni y son None ', z)

# Un ejemplo de uso de is / ==
x= 1.j
y= 1.j
print()
print('valor de x', x)
print('valor de y', y)
print('x==y', x==y)
print('x is y', x is y)


# Un ejemplo de uso de in
palabra= 'abracadabra'
print(palabra, 'contiene una o: ', 'o' in palabra)
print(palabra, 'contiene una a: ', 'a' in palabra)
print(palabra, 'contiene una ca: ', 'ca' in palabra)
print(palabra, 'contiene una co: ', 'co' in palabra)
print(palabra, 'NO contiene una i: ', 'i' not in palabra)

````

## Gestión de excepciones

Al igual que otros lenguajes orientados a objetos, Python permite gestionar excepciones de errores de ejecución. Para ello, se contempla la sentencia ```try..except''

Un ejemplo:

````python
a= 1
b= 0

try:
  c= a/b
except:
  print('Ha ocurrido una excepción')
````

Se pueden capturar excepciones de varios tipos y tratarlas por separado. Para ello, sólo es necesario indicar el nombre de la excepción tras el uso de ```except```. Las excepciones que se incorporan de forma nativa en Python se pueden ver en el enlace <a href="https://docs.python.org/3/library/exceptions.html">https://docs.python.org/3/library/exceptions.html</a>


También se nos proporciona la posibilidad de ejecutar otro código en caso de que no existan exepceiones mediante ```try..except..else```:

````python
# Ejemplo de captura de varias excepciones
try:
  print(variable_no_definida)
except NameError: # Error de nombre no definido
  print("La Variable  'variable_no_definida' no está definida")
except:
  print("Se ha capturado otra excepción")


# Ejemplo de ejecución de código una vez que no se ha producido una excepción
variable_definida= None
try:
  print(variable_definida)
except NameError: # Error de nombre no definido
  print("La Variable  'variable_definida' no está definida")
else:
  print('Se ha cmostrado el valor de "variable_definida"')
````

Por último, también se permite ejecutar un código tanto si ha habido una excepción como si no, mediante ```try..except..finally

````python
a= 1
b= 0

try:
  c= a/b
except ZeroDivisionError: # Captura de excepción de división por cero
  print('Ha ocurrido una excepción')
finally:
  c= 0

# El valor de c siempre será cero indepenndientemente de los valores que pongamos a a y b
print('valor final de c: ', c)
````

# **Sentencias condicionales**

```
if condicion:
  sentencia1
  sentencia2
  ...
```

La sintaxis de ```if..else```:


```
if condicion:
  sentenciaIf1
  sentenciaIf2
  ...
else:
  sentenciaElse1
  sentenciaEkse2
  ...
```

Python también permite anidar varios *ifs* sin necesidad de tabular mediante expresiones de tipo ```if..elif..else```. Su sintaxis es:

```
if condicion1:
  sentenciaIf1
  sentenciaIf2
  ...
elif condicion2:
  sentenciaElIf1
  sentenciaElIf2
  ...
else:
  sentenciaElse1
  sentenciaEkse2
  ...
```

**NO OLVIDAR LA IMPORTANCIA DE LA INDENTACIÓN, ASÍ COMO LA TERMINACIÓN CON EL CARACTER (:) TRAS LA ESCRITURA DE LAS CONDICIONES DE CADA IF/ELIF/ELSE**.

<u><b>Una mala indentación o el olvido de los : suele ser una fuente de errores muy común</b></u>.



Veamos algunos ejemplos:

````python

a= None

# Ejemplo de if..else
if a is not None:
  print('a no es None')
else:
  print('a es None')


# Ejemplo de if..elif
a= 3
b= 4
c= 5
maximo= None

if a>b and a>c:
  maximo= a
elif b>a and b>c:
  maximo= b
elif c>a and c>b:
  maximo= c

if maximo is None:
  print('No hay ningún elemento más grande que el resto')
else:
  print('El elemento {} es mayor que el resto'.format(maximo))


````

## Sentencia *while*


La sintaxis del bucle while es la siguiente:

```

  sentencia1

  ...
SntenciaFueraDelWhile
```

Las sentencias indentadas se ejecutarán mientras que la condición dada se cumpla. Una vez que dicha condición devuelva el valor lógico **False**, el programa continuará por ```SentenciaFueraDelWhile``` en adelante.

Veamos un ejemplo:

````python
# Ejemplo de bucle while

debug= True # Poner esta variable a Fase para que el ejemplo tome valores desde línea de comandos

# El bucle para cuando a sea un valor entre 1 y 10
a= -1
while a<=0 or a>10:

  print('Dime un valor de 1 a 10: ')
  if debug:
    a= 7
  else:
    try:
      a= int( input() )
    except:
      print('No seas malote. Introduce algo que pueda convertirse a entero.')
      a= -1
print('Has introducido el valor {}'.format(a))
````

## Sentencia *for*

La sentencia ```for``` de Python puede resultar un poco diferente a programadores acostumbrados a lenguajes como C/C++. Se utiliza para iterar sobre un conjunto de elementos, y su sintaxis es:

```
for valor in coleccion_de_valores:
  sentencia1
  sentencia2

```

## Funciones útiles: reversed, zip y enumerate



Las funciones ```reversed```, ```enumerate``` y ```zip``` son muy útiles a la hora de simplificar código Python:

- ```reversed``` se utiliza para iterar una secuencia en orden inverso.
- ```enumerate``` sirve para indexar elementos dentro de una colección a iterar. Supongamos que iteramos sobre una secuencia S= [a, b, c, d, ...]. Entonces, en cada iteración, enumerate(S) devuelve dos valores i, v:
  - ```i``` : El índice o **contador** (comenzando desde 0) del número de elementos visitados/iterados en la colección.
  - ```v``` : El valor del i-ésimo elemento visitado.
- ```zip``` se utiliza para agrupar dos o más colecciones dentro de un mismo iterador; es decir, se puede usar para iterar sobre dos o más colecciones simultáneamente. Sean S1= [s11, s12, s13, s14, ...] y S2= [s21, s22, s23, s24, ...] dos colecciones **del mismo tamaño**. Entonces ```zip(S1, S2)``` devuelve un iterador que permite viajar sobre las dos colecciones devolviendo en cada iteración dos valores s1,s2:
  - ```s1```: El i-ésimo elemento de la colección S1 visitado.
  - ```s2```: El i-ésimo elemento de la colección S2 visitado.

Veamos algunos ejemplos sobre cómo utilizar estas funciones:

````python
# Ejemplos de reversed:
cadena1= 'hola'
cadena2= ''

for v in reversed(cadena):
  cadena2+= v
print('{} es {}, pero al revés'.format(cadena2, cadena1))

# OJO: Python es muy lento. Debemos conocer todos sus entresijos para no ejecutar código ineficiente.
# Por ejemplo, el código siguiente equivale al anterior:
cadena2= cadena[::-1]
print('{} es {}, pero al revés'.format(cadena2, cadena1))
````

````python
# Ejemplos de enumerate
cadena= 'hola'

for i, v in enumerate(cadena):
  print('El elemento de la posicion {} de {} es {}'.format(i, cadena, v))
````

````python
# Ejemplos de zip
cadena1= '5678'
cadena2= 'hola'

# Ejemplo de zip
print('Ejemplo 1: Iteración sobre dos cadenas {}  y {}'.format(cadena1, cadena2))
for v1, v2 in zip(cadena1, cadena2):
  print('Obtengo valores {} y {} desde {} y {}, respectivamente.'.format(v1, v2, cadena1, cadena2))



# Ejemplo de combinación de enumerate y zip
# OJO: Enumerate devuelve dos valores i, v
#      A su vez, ZIP devuelve dos valores v1, v2.
#      Se debe indicar que el valor v=(v1, v2) entre paréntesis
print('Ejemplo 2: Combinación de enumerate con zip')
for i, (v1, v2) in enumerate(zip(cadena1, cadena2)):
  print('El {}-esimo elemento de cada cadena {} y {} es {} y {}, respectivamente.'.format(i, cadena1, cadena2, v1, v2))
````

## Listas



Una lista **L** (tipo```list```) es una secuencia de elementos indexada entre ```0``` y ```len(L)-1```, ambos incluidos. Una lista es un objeto **mutable**; es decir, que se puede modificar. Dicha modificación puede consistir en:

- Añadir nuevos elementos a la lista (métodos ```append(elm)```, ```extend(otra_lista)```, ...)
- Eliminar elementos de la lista (métodos ```remove(elm)```, ```pop(indice)```, ```clear()```, ...)
- Acceso a elementos en el índice ```idx``` : ```L[idx]```
  - Caso particular: Acceso al último elemento como ```L[-1]```. Al penúltimo como ```L[-2]```, etc.
- Acceso a subconjunto de elementos por rango ```L[inicio:fin:pasos]```
   - ```L[0:k]``` : Lista con los elementos desde el índice 0 hasta el k-1.
   - ```L[:k]``` : Lista con los elementos desde el índice 0 hasta el k-1.
   - ```L[k:]``` : Lista con los elementos desde el índice k hasta el len(L)-1.
   - ```L[0:k:2]``` : Lista con los elementos ```L[0], L[2], L[4], ..., L[k-1]``` (si k es par)

Creación de listas:

- Lista vacía: ```L= []```
- Lista con elementos: ```L= [1, 2, 3, 4]```
- Lista con el mismo elemento **(OBJETO)** n veces: ```L= [8]*n```
- Lista desde elementos iterados

Las operaciones con listas se pueden ver en el enlace <a href="https://www.w3schools.com/python/python_lists.asp">https://www.w3schools.com/python/python_lists.asp</a>

La siguiente celda muestra varios ejemplos para acceder a elementos de una lista.

````python
# Lista con 10 elementos dew tipo entero
L= [4, 98, 10, 5, 67, 23, 43, 88, 44, 66]
print('La lista L: ', L)
print('Su primer elemento: ', L[0])
print('Su último elemento: ', L[-1], L[len(L)-1])
print('Elementos desde el índice 2 al 5 (no incluido)', L[2:5])
print('Elementos desde el índice 2 al 10 de 3 en 3: ', L[2:10:3])
print('Todos los elementos de la lista:', L[:])
print('Lista invertida: ', L[::-1])

````

**CUIDADO CON EL USO DE \* PARA CREAR LISTAS INICIALIZADAS CON EL MISMO VALOR. RECORDEMOS QUE TODO PUEDE VERSE COMO UN OBJETO Y COMO REFERENCIA AL MISMO**:

````python

### ESTE CÓDIGO NO DA PROBLEMAS

# Lista de 10 elementos inicializada a 0
L= [0]*10
print('Lista inicializada: ', L)

# Modificamos algún elemento de la lista:
L[2]= 7
print('Lista modificada OK: ', L)


### ESTE CÓDIGO SI DA PROBLEMAS

# Lista de 10 listas inicializadas con un elemento 0
L= [ [0] ]*10
print('Lista inicializada: ', L)

# Modificamos el elemento 0 de la primera lista
L[0][0]= 7
print('Lista modificada MAL: ', L)
````

En el último ejemplo de la celda anterior, todas las sublistas de la lista han modificado su primer elemento de 7 a 0. **¿Porqué?**:

No olvidemos que **casi todo son punteros/referencias** en Python. La sentencia

```L= [ [0] ]*10```

primero crea un objeto lista inicializado con un elemento 0, y devuelve dicho objeto (puntero/referencia al mismo). Luego se crea una lista de 10 elementos donde cada uno de ellos **es una referencia** al objeto ```[0]``` creado. Al hacer ```L[0]``` estamos accediendo al primer elemento de la lista L, que es **una refeerncia al objeto**. Al hacer L[0][0] estamos accediendo al primer elemento de dicho objeto, y lo modificamos con un 7. Tanto L[0] como L[1], como L[2], etc., apuntan al mismo objeto. Es normal que la lista de 10 referencias al mismo objeto muestre lo mismo: el valor 7.

### Algunos métodos útiles de listas

Algunas de las funcionalidades más usadas en listas se corresponden con la inserción de nuevos elementos, su borrado, búsqueda u ordenación. En este cuaderno mostramos los siguientes:

- Función ```len(L)```: Devuelve el número de elementos de la lista ```L```
- Método ```append(elm)```: Inserta el elemento ```elm``` al final de la lista.
- Método ```extend(L2)```: Concatena la lista actual con la lista ```L2``` pasada como argumento.
- Método ```pop(idx)```: Elimina el elemento localizado en la posición ```idx``` de la lista.
- Método ```remove(elm)```: Elimina la primera ocurrencia del elemento ```elm``` en la lista
- Método ```clear()```: Elimina todos los elementos de la lista
- Método ```index(elm)```: Devuelve el primer índice de la lista que contiene el elemento ```elm```
- Método ```sort()```: Ordena la lista (por defecto, de forma ascendente)

Veamos algunos ejemplos en el siguiente código:


````python
L= [] # Creación de lista vacía
print('Lista L inicial: {}'.format(L))

# Inserción de elementos del 1 al 10, por duplicado y en orden inverso
for i in range(10, 0, -1):
  L.append(i)
  L.append(i)
print('Lista L tras append: {}'.format(L))

# Ordenación de la lista
L.sort()
print('Lista L tras sort: {}'.format(L))

# Eliminación del elemento 2 (SÓLO PRIMERA OCURRENCIA)
L.remove(2)
print('Lista L tras remove: {}'.format(L))

# Eliminación del último elemento de la lista
L.pop()
print('Lista L tras pop: {}'.format(L))

# Eliminación del primer elemento de la lista
L.pop(0)
print('Lista L tras pop(0): {}'.format(L))

# Uso de len
print('La lista {} tiene {} elementos.'.format(L, len(L)))

# Creación de lista concatenando L consigo misma
L.extend(L)
print('Lista L tras extend: {}, ahora con {} elementos.'.format(L, len(L)))


# Búsqueda del elemento 3 (PRIMERA OCURRENCIA)
valor= 3
i= L.index(valor)
print('El elemento {} se encuentra en la posición {} de la lista {}.'.format(valor, i, L))

# Limpieza de la lista
L.clear()
print('Lista L tras clear: {}, ahora con {} elementos.'.format(L, len(L)))

````

## Tuplas

Una tupla (tipo ```tuple```) es un registro que contiene un número fijo de elementos. Las tuplas son **objetos inmutables** en Python de modo que, una vez creada una tupla, esta no podrá modificarse **aunque sí el valor de sus elementos si estos son objetos**.

Mientras que el tipo ```list``` define listas utilizando los corchetes (por ejemplo en la lista ```[1, 2, 3]```), el tipo ```tuple``` utiliza **paréntesis** (por ejemplo en la tupla de 3 elementos ```(1, 2, 3)```).

Las tuplas también tienen disponible el operador sobre colecciones **len()**, de modo que podemos conocer el tamaño de una tupla ```T``` (número de elementos en la misma) mediante ```len(T)```. Además, cada elemento se accede con un índice entre ```T[0]``` y ```T[len(T)-1]```.

Veamos algunos ejemplos en la siguiente celda de código:

````python

# Creación de lista con los elementos 10, 20, 30
L= [10, 20, 30] # OJO: Las tuplas usan corchetes

# Creación de tupla con los elementos 10, 20, 30
T= (10, 20, 30) # OJO: Las tuplas usan paréntesis
print('La lista L vale {} y la tupla T vale {}'.format(L, T))

# Modificamos el segundo elemento de la lista
L[1]= 40

# Modificamos el segundo elemento de la tupla
try:
  T[1]= 40
except:
  print('Upssss. Las tuplas no se pueden modificar')

print('Tras la modificación: La lista L vale {} y la tupla T vale {}'.format(L, T))

# Creación de tupla con 3 listas L
T2= (L, L, L)
try:
  T2[0].append(40) # Insertamos el elemento 40 en la lista de la posición 0
except:
  print('Upssss. Las tuplas no se pueden modificar')
else:
  print('Que noooooo, que los elementos sí que se pueden modificar!')

print('Tupla T2 tras la modificación: {}'.format(T2))
````

## Conjuntos

Un conjunto (tipo```set```) es una **colección de elementos sin repetir, y sin un orden establecido a priori**. Los conjuntos son **objetos mutables** y permiten un acceso rápido a cada elemento en la colección, sobre todo si deseamos comprobar si un elemento dado está incluido o no dentro de dicha colección.

La creación de conjuntos en Python se puede realizar de diversas formas:

- Conjunto vacío: ```a= set()```
- Conjunto con elementos: ```a= {1, 2, 3}```
- Conjunto con elementos desde iterador: ```a= {x for x in range(1, 4)}```

Dado que los elementos de un conjunto no tienen orden establecido, no se puede acceder a cada elemento con un índice (por ejemplo ```a[0]```), como sí ocurre con las listas y las tuplas.

Las operaciones entre conjuntos y elementos más comunes son las siguientes:

- Inserción de un nuevo elemento ```elm``` en el conjunto ```S```: ```S.add(elm)```
- Borrado de un elemento ```elm``` en el conjunto ```S```: ```S.remove(elm)```
- Limpiar el conjunto ```S``` de todos sus elementos: ```S.clear()```
- Comprobar si un elemento ```elm``` está en un conjunto ```S```: ```elm in S```, con resultado **True** o ***False***.

Las operaciones entre conjuntos son las típicas de la teoría de conjuntos:

- Intersección $S_1\cap S_2$ de dos conjuntos $S_1$ y $S_2$: Devuelve un conjunto con los elementos comunes en $S_1$ y $S_2$, usando el método ```S1.intersection(S2)```.
- Unión $S_1\cup S_2$ de dos conjuntos $S_1$ y $S_2$ (**sin repetir elementos**), usando el método ```S1.union(S2)```
- Diferencia $S_1\backslash S_2$ del conjunto $S_1$ menos $S_2$: Devuelve un conjunto con todos los elementos de $S_1$ que no estén en $S_2$, mediante el método ```S1.difference(S2)```.

Veamos algunos ejemplos:


````python
C= {1, 2, 3} # Conjunto con 3 elementos. OJO: Son llaves
print('Conjunto C inicial: {}'.format(C))

# Inserción del -1
C.add(-1)
print('Conjunto C tras insertar -1: {}'.format(C))

# Borrado del 1
C.remove(1)
print('Conjunto C tras eliminar 1: {}'.format(C))

# Nuevo conjunto con elementos del 0 al 9
C2= set()
for i in range(10):
  C2.add(i)
print('Nuevo conjunto C2: {}'.format(C2))

# Otra forma alternativa:
C2= {x for x in range(10)}
print('Nuevo conjunto C2: {}'.format(C2))

# Intersección de C y C2
I= C.intersection(C2)
print('Intersección de {} y {}: {}'.format(C, C2, I))

# Unión de C y C2
U= C.union(C2)
print('Unión de {} y {}: {}'.format(C, C2, U))

# Diferencia de C y C2
D1= C.difference(C2)
print('Diferencia de {} y {}: {}'.format(C, C2, D1))

# Diferencia de C2 y C
D2= C2.difference(C)
print('Diferencia de {} y {}: {}'.format(C2, C, D2))


# Comprobación de si un elemento pertenece a un conjunto o no
value= 1
if value in C2:
  print('El valor {} pertenece a {}'.format(value, C2))
else:
  print('El valor {} NO pertenece a {}'.format(value, C2))
````

## Diccionarios

Un diccionario (tipo ```dict```) permite asociar pares de objetos *clave-valor*, como por ejemplo ocurre en una tabla Hash. Se utiliza comúnmente para acceder rápidamente a valores dada una clave asociada a los mismos.

Python también utiliza las llaves {} para representar diccionarios aunque, en este caso, la definición se realiza indicando el par *clave:valor* separado por el símbolo de dos puntos.

Algunas formas de crear diccionarios en Python son:

- Diccionario vacío: ```a= {}```
- Diccionario con clave/valor: ```a= {clave1 : valor1, clave2 : valor2, ...}```

Algunas operaciones comunes:

- Se puede comprobar si una clave ```k`` está en un diccionario ```D``` mediante el operador ```in```. Por ejemplo: ``` k in D ```
- Se puede acceder al valor de una clave usando corchetes: ```D[k]```
- Se puede añadir un par (clave,valor), o modificar el valor de una clave, simplemente realizando la asignación ```D[k]= valor```.
- Se puede acceder a todos los pares (clave,valor) de un diccionario con el método ``ìtems()```. Por ejemplo: ```todos_items= D.items()```
- Se puede borrar el contenido de un diccionario con ```D.clear()```

Más información sobre diccionarios en Python en el enlace <a href="https://www.w3schools.com/python/python_dictionaries.asp">https://www.w3schools.com/python/python_dictionaries.asp</a>

Veamos un ejemplo de creación y uso de diccionarios en la siguiente celda de código:

````python
# Ejemplo de uso de un diccionario
palabra= 'tegucigalpa'

# Diccionario vacío
D= {}

# Creamos un diccionario de claves caracteres y valores listas de enteros
# Servirá para almacenar en qué posiciones de la palabra anterior aparece
# cada caracter
for posicion, caracter in enumerate(palabra): # Recorremos todos los caracteres de la palabra

  # Comprobamos si la clave (caracter) existe.
  # De no ser así, le asignamos una lista vacía.
  if caracter not in D:
    D[caracter]= []

  # Insertamos posición para el caracter
  D[caracter].append(posicion)

print('La palabra {} (longitud={}) tiene {} caracteres distintos.'.format(palabra, len(palabra), len(D)))

# Recorremos todos sus caracteres (claves)
for key in D:
  print('El caracter {} aparece en las posiciones {}'.format(key, D[key]))
````

# **Funciones y procedimientos**

## Definición de funciones.

Los módulos en Python (funciones y procedimientos) se definen con la palabra reservada ```def```. Para cada módulo es necesario indicar un *nombre del módulo* y, entre paréntesis y separados por comas, el *nombre de sus argumentos* (en caso de que tenga alguno).

La sintaxis general para creación de módulos es la siguiente:

```
def nombre_modulo([param1, param2, param3, ...]):
   Sentencia1
   Sentencia2
   ...
```

**OJO: Al final de la línea del ```def``` hay que poner dos puntos**. En la sintaxis anterior, los corchetes indican que los parámetros de entrada son opcionales. También es posible dar valores por defecto a algunos parámetros; no obstante, una vez que se da un valor por defecto a uno de ellos, **todos los siguientes también deberán tener**. Por ejemplo:

```
def nombre_modulo([param1, param2= 0, param3= None, param4= 'cadena', ...]):
   Sentencia1

```

**La ejecución de un módulo requiere que se pasen valores a todos sus parámetros que, al menos, no tengan valores por defecto.** Además, se puede indicar concretamente a qué parámetro pertenece cada valor indicando su nombre y valor asignado en la llamada.

La siguiente celda de código muestra algunos ejemplos de procedimientos y llamadas a los mismos:

````python
# Definición de procedimiento sin parámetros
def print_hola():
  print('hola')

# Definición de procedimiento con 3 parámetros (el último por defecto a 2025)
def fecha_de_hoy(dia, mes, anio=2025):
  print('la fecha de hoy es: {}/{}/{}'.format(dia, mes, anio))


# Uso de procedimientos sin nombramiento de parámetros
print_hola() # Llamada al procedimiento
fecha_de_hoy(10, 12) # 10/12/2025
fecha_de_hoy(10, 12, 2026) # 10/12/2026

# Uso de procedimientos nombrando parámetros en un orden arbitrario
fecha_de_hoy(mes= 1, dia= 3) # Fijémonos en que el orden de los parámetros no es el definido en la función, pero no hay problema porque se indica su nombre
fecha_de_hoy(mes= 1, anio= 2027, dia= 3)
````

### Salidas de una función. La palabra reservada *return*

Como norma general, utilizamos las funciones para implementar sub-programas que realicen cálculos y devuelvan resultados. Al igual que en otros lenguajes como C/C++, estos resultados se devuelven con la palabra reservada **return**.

Veamos un ejemplo en la siguiente celda de código:

````python

# Función que devuelve el factorial de un número de entrada n
def factorial(n):

  # Comprobación de errores en la siguiente línea:
  # assert lanza una excepción si la condición interna no se cumple
  # isinstance comprueba que el primer argumento es del tipo del segundo
  assert( isinstance(n, int) and n>=0 )

  if n<=1:
    return 1

  f= 1
  for i in range(2, n+1):
    f*= i
  return f


# Ejemplo de uso de la función anterior
for i in range(-1, 4):
  try:
    resultado= factorial(i)
  except:
    print('No se puede calcular {}!'.format(i))
  else:
    print('{}!= {}'.format(i, resultado))
````

**¿Qué ocurre si la función debe devolver más de un valor?** Como Python incorpora de forma nativa las tuplas, se pueden devolver varios valores separados por comas.

Por ejemplo, veamos una función que calcula el cociente y el resto de una división entre números naturales:

````python

# Función que calcula el cociente y el resto de una división natural
def divide(dividendo, divisor):

  # ATENCIÓN: Las líneas muy largas se pueden dividir con el caracter \, continuando en la siguiente línea
  assert(isinstance(dividendo, int) and isinstance(divisor, int) \
         and dividendo>=0 and divisor > 0)

  c= dividendo // divisor
  r= dividendo % divisor

  return c, r # Devolvemos cociente y resto


# Ejemplo de uso
cociente, resto= divide(dividendo= 7, divisor= 3)
print('Al dividir 7 entre 3 obtenemos el cociente {} y el resto {}.'.format(cociente, resto))
````

### Salidas de una función. ¿Paso por referencia?

Como regla general, en Python no existe el paso por referencia. No se pueden modificar los valores de un parámetro dado como argumento. Veamos un ejemplo:

````python
# Procedimiento que pretende limpiar una lista
def limpiar_lista_mal(lista):
  lista= [] # Limpiamos la lista simplemente asignándola a la lista vacía

# Veamos un ejemplo
mi_lista= [1, 2, 3]
print('Inicialmente, mi_lista vale {}'.format(mi_lista))
limpiar_lista_mal(mi_lista)
print('Al finalizar, mi_lista vale {}'.format(mi_lista))
````

<b><u>Sin embargo</u></b>, recordemos que:

- Casi todo en Python son referencias.
- Casi todo en Python son objetos.

Esto implica que dentro de una función, si utilizamos métodos de un objeto, este **sí se verá modificado en el exterior de la función si es el caso**.

Veamos un ejemplo:

````python
# Función para limpiar una lista
def limpiar_lista_bien(lista):
  assert(isinstance(lista, list))
  lista.clear()

# Función para insertar un elemento en una lista
def insertar_en_lista(elemento, lista):
  lista.append(elemento)


# Veamos un ejemplo
mi_lista= [1, 2, 3]
print('Inicialmente, mi_lista vale {}'.format(mi_lista))
limpiar_lista_bien(mi_lista)
print('Al finalizar limpiar_lista_bien, mi_lista vale {}'.format(mi_lista))

insertar_en_lista(lista=mi_lista, elemento= 'oye')
insertar_en_lista(lista=mi_lista, elemento= 'tú')
print('Al finalizar insertar_en_lista, mi_lista vale {}'.format(mi_lista))
````

### Iteradores: La palabra reservada ```yield```

Todos sabemos que la palabra reservada ```return```, ejecutada dentro de una función, finaliza la misma y destruye el entorno de ejecución de dicha función.

La palabra reservada ```yield``` permite finalizar la ejecución de una función, **pero sin destruir su entorno de ejecución**. De este modo, la próxima vez que realicemos una llamada a la función que ha devuelto un valor con yield, continuará ejecutándose desde ese punto.

```yield``` se puede utilizar, por ejemplo, para la creación de iteradores. Veamos un ejemplo:

````python
# Función que se ejecuta un máximo de max_veces (por defecto 10)
def mi_iterador(max_veces=10):
  i= 0
  while i < max_veces:
    print('Ejecución de mi_iterador con valor i={}'.format(i))
    yield i # Salida de la función, pero no destrucción de su entorno
    i= i+1 # La siguiente vez que se ejecute continuará desde aquí

  return None # Termina la función con return para cerrar su entorno de ejecución. Valor None si la usamos como iterador normal

# Ejemplo de uso
for var in mi_iterador(): # termina cuando obtiene el valor None
  pass

# Comienzo de nuevo de la ejecución desde el principio
var= mi_iterador()
print('var es un iterador/generador de datos, como podemos ver: ', var)
````

En el formato anterior, los corchetes delimitan zonas del código opcionales.

**Toda función lambda debe devolver algún valor (None como mínimo)**

````python

# Función lambda que crea una lista de enteros entre inicio y fin, ambos incluidos
crear_lista= lambda inicio, fin : [x for x in range(inicio, fin+1)]

# Ejemplo de uso
L= crear_lista(3, 5)
print('La lista L vale {}'.format(L))


# Otro ejemplo: Función sigmoide
from math import exp
sigmoid= lambda x : 1/(1+exp(-x))

# Ejemplo de uso
value= 0
print('La función sigmoide evaluada en {} tiene el valor {}'.format(value, sigmoid(value)))

````

# **Orientación a objetos**

Python es también un lenguaje orientado a objetos. Podemos construir clases que después instanciemos como objetos del tipo construido. Se crean con la palabra reservada **class**:

````python
# Creación de clase vacía ParDeNumeros
class ParDeNumeros:
  pass
````

Debemos tener en cuenta los siguientes aspectos:

- Todos los métodos de la clase deben estar **correctamente indentados**.
- Todos los métodos a usar por los objetos de la clase deben tener un primer argumento denominado **self**.
- Los métodos estáticos (de la clase) deben estar indentados, **pero no deben incorporar el primer argumento self**.
- Sólo puede haber un **constructor**. Este es el método ```def __init__(self, [parametros])```
- Se puede **sobrecargar operadores** (ver pequeño ejemplo introductorio en <a href="https://www.geeksforgeeks.org/operator-overloading-in-python/">https://www.geeksforgeeks.org/operator-overloading-in-python/</a>
- Los **atributos del objeto** se pueden crear en cualquier método. Sólo es necesario que el nombre del atributo vaya precedido por ```self.NombreDeAtributo```.
- El nombre de los métodos y atributos **privados** deben comendar por doble subrayado ```self.__atributo_privado```.

Un ejemplo:

````python

# Creación de clase vacía ParDeNumeros
class ParDeNumeros:

  # Constructor con dos parámetros x,y
  def __init__(self, x, y):
    self.x= x # Atributo x del objeto= argumento x
    self.__y= y # Atributo y del objeto= argumento y . self.__y es privado


  # Método estático para crear el par (0,0)
  def ParCero():
    return ParDeNumeros(0,0)

  # Método para asignar los valores del par
  def set(self, x, y):
    self.x= x
    self.__y= y

  # Método para devolver la componente X
  def getX(self):
    return self.x

  # Método para devolver la componente Y
  def getY(self):
    return self.__y

  # Sobrecarga de str(objeto)
  def __str__(self):
    return '('+str(self.x)+','+str(self.__y)+')'

  # Sobrecarga del operador de comparación ==
  def __eq__(self, other):
    return self.x == other.x and self.__y==other.__y
````

Ejemplo de uso de la clase:

````python

# Uso de métodos estáticos
cero= ParDeNumeros.ParCero()
print('El par cero es '+str(cero))

# Uso del constructor
a= ParDeNumeros(1,2)
b= ParDeNumeros(2, 3)

print('a vale '+str(a)+' y b vale {}'.format(b))
print('La componente X de a es: {} / {}'.format(a.getX(), a.x))

print('La componente Y de a es: {}'.format(a.getY()))

try:
  print('La componente Y de a es: {}'.format(a.__y))
except:
  print('El atributo a.__y es privado y no accesible')

print('¿a==b? {}'.format(a==b))
b.set(1, 2)
print('Ahora he asignado b a '+str(b))
print('¿a==b? {}'.format(a==b))
````

## Herencia

Python soporta la herencia, y también la sobrecarga de métodos en clases heredadas. La clase padre debe indicarse a la hora de definir la clase heredada:


````python
class Complejo(ParDeNumeros): # Clase Complejo extendida desde ParDeNumeros
  pass
````

A la hora de crear un objeto de la clase heredada, nosotros somos responsables de llamar (o no) a los constructores de la/s clase/s padre que deseemos, utilizando la palabra reservada **super**.

````python
class Complejo(ParDeNumeros): # Clase Complejo extendida desde ParDeNumeros

  # Constructor
  def __init__(self, real, imag):
    super().__init__(x= real, y= imag)


  # Sobrecarga de str
  def __str__(self):
    txt= str(self.x) # Accesible porque no es privada.

    if self.getY()>=0 : # self.__y no es accesible porque es privada de ParDeNumeros
      txt+= '+'

    txt+= str(self.getY())+'j'

    return txt

  # Devuelve un complejo conjugado del actual
  def conjugate(self):
    return Complejo(self.getX(), -self.getY())

  # Sobrecarga de la suma
  def __add__(self, other):
    return Complejo(self.getX()+other.getX(), self.getY()+other.getY())
````

Ejemplo de uso:

````python

a= Complejo(1, 2)
b= Complejo(2, 3)

print('El complejo a={} y el b={}'.format(a,b))
print('El conjugado de {} es {}'.format(a, a.conjugate()))
print('La suma de {}+{} es {}'.format(a, b, a+b))
````
