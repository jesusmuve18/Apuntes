<b><h1>Ejemplo de uso de bibliotecas externas</h1></b>

Las bibliotecas, en Python, se estructuran en paquetes.

Podemos incluir una biblioteca Python externa de varias formas:

<ul>
<li>Haciendo uso de <b>import</b> y el nombre de la biblioteca</li>
<li>Haciendo uso de <b>import</b>, el nombre de la biblioteca y un <b>alias</b></li>
</ul>

```python
import matplotlib # Inclusión de la biblioteca matplotlib
import matplotlib as mpl # Inclusión de la biblioteca matplotlib con alias
```

También se pueden incluir <b>paquetes</b> de una biblioteca si no queremos toda su funcionalidad.

Un paquete se carga <b>desde</b> una biblioteca con <b>>from</b> o, directamente, accediendo al nombre del paquete de la biblioteca separando su nombre por puntos.

```python
from matplotlib import pyplot
import matplotlib.pyplot
```

La carga de paquetes también admite alias

```python
from matplotlib import pyplot as plt
```

<b><h1>Ejemplo rápido de generación de gráficas y su visualización en el cuaderno</h1></b>

Veamos ahora un ejemplo de generación de gráficas con el paquete PyPlot de MatplotLib.
