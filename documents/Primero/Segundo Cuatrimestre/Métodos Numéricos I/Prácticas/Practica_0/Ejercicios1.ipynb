# Ejercicio 1

Escribe un pequeño programa que calcule el máximo entre 4 números que debe introducir el usuario sin utilizar la función max de Python

```python
DEBUG=0 # variable utilizada para hacer un DEBUG
        # si Debug=1 se mostrarán los pasos que se siguen en la ejecución del programa
        # si Debug=0 (o cualquier otro valor) no se mostrarán los carteles

cant=4   # cantidad de números que se leen para calcular el máximo (afecta al número de peticiones)
lista=[] # almacena los números que se leen
    
for x in range (cant):
    lista.append(int(input(f'Valor {x+1}: '))) 

maximo=lista[0]

for x in range (1,cant):
    
    if (DEBUG == 1):
        print()    # inserta una línea en blanco
        print(f'Comparo {lista[x]} con {maximo}')
    
    if lista[x]>maximo:
        maximo=lista[x]
        if (DEBUG == 1):
            print('Es mayor')
    else:
        if (DEBUG == 1):
            print('Es menor')
            

print()     # inserta una línea en blanco
print('El maximo es ', maximo)  # muestra el valor guardado en la variable 'm'

lista.clear()   # vacía la lista para evitar errores en la siguiente ejecución
    
```

# Ejercicio 2

Crea un array de `Numpy` de 20 filas y 10 columnas de manera que el elemento $(i,j)$ sea  $\frac{1}{i^2+j-1}$ para $1\le i\le 20$ y $1\le j\le10$. Ten en cuenta que las filas y columnas se numeran a partir de 1 pero en Python la numeración empieza en 0. El primer elemento del array debe ser 1.

```python
import numpy as np   # incluyo la biblioteca de Numpy

def imprime(matriz , n_decimales=3 , separador = '    '): #función que imprime una matriz bidimensional
    
    cad='{0:.'
    cad+=str(n_decimales)
    cad+=('f}')
    
    print('[', end='')
    
    for i in range (matriz.shape[0]-1):
        
        if(i==0):
            print('[', end='')
        else:
            print(' [', end='')
        
        for j in range (matriz.shape[1]-1):
            
            if(j==matriz.shape[1]-2):
                print(cad.format(matriz[i][j],n_decimales), end='') 
                
            else:
                print(cad.format(matriz[i][j],n_decimales), end=separador)
                
        if(i==matriz.shape[0]-2):
            print(']]')
                
        else:
            print(']')



matrizA=np.array([[(1/((i+1)**2+(j+1)-1)) for j in range (0,11)]for i in range (0,21)])

    
imprime(matrizA , 5 , '     ')
```

# Ejercicio 4

Crea ahora un array $B$ de números aleatorios entre 0 y 1 de tamaño $5\times 5$. Calcula ${(B+2I)}^{8}$.

```python
import numpy as np

matrizB=np.random.rand(5,5)
print('Matriz generada:')
print(matrizB)

ide=np.eye(5) #matriz identidad

resultado=np.linalg.matrix_power((matrizB + (2*ide)),2)

print()
print('Matriz (B+2I)^10:')
print(resultado)
```
