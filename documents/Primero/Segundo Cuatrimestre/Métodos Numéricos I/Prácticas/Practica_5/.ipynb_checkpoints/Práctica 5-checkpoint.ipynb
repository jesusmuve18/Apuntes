```python
import numpy as np
import matplotlib.pyplot as plt

# FEN√ìMENO DE RUNGE

n1=18
definicion=np.linspace(-5,5,n1)

n2=100
puntos=np.linspace(-5,5,n2)

n3=100
xp=np.linspace(-5,5,n3)


def f(x):
    return (1/(1+x**2))

yp=f(xp)
imagenes=f(puntos)

def polinomio (puntos,x):
    n=len(puntos)-1;
    if (len(puntos)==1):
        return f(puntos)
    else:
        puntos1=np.delete(puntos,[0])
        puntos2=np.delete(puntos,[n])
        return (((x-puntos[0])*polinomio(puntos1,x)-(x-puntos[n])*polinomio(puntos2,x))/(puntos[n]-puntos[0]))


plt.plot(puntos,polinomio(definicion,puntos),'r',label='Polinomio de interpolaci√≥n')
#plt.plot(puntos,polinomio(definicion,puntos),'g.',label='$1/(1+x^2)$')

plt.plot(xp,f(xp),'k',label='$(1+x^2)^{-1}$')

plt.xlabel('x')  #dibuja la etiqueta del eje x con el contenido de ''
plt.ylabel('y')  #dibuja la etiqueta del eje y con el contenido de ''
plt.grid(color='lightblue') #dibuja la cuadr√≠cula de fondo
plt.title('Polinomio de interpolaci√≥n',color='darkblue', loc='center') #muestra el t√≠tulo del gr√°fico
plt.legend() #muestra las etiquetas (label) de las funciones
plt.ylim(-1,2) #establece el l√≠mite vertical de visi√≥n en la gr√°fica
plt.xlim(-6,6) #establece el l√≠mite horizontal de visi√≥n en la gr√°fica
plt.plot([0,0],[-1,2],'dimgrey') #dibuja el eje Y
plt.plot([-6,6],[0,0],'dimgrey') #dibuja el eje X
plt.rcParams["figure.figsize"] = (10, 10) #fija el tama√±o de la gr√°fica (dejar en 10,10)
plt.show()
```
