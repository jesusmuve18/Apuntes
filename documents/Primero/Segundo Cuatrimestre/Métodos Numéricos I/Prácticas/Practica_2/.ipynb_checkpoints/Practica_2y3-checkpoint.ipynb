**Ejercicio 1**: Escribe una función que resuelva por sustitución progresiva un sistema en el que la matriz de coeficientes es triangular inferior. Comprueba que se obtiene el mismo resultado utilizando dicha función que mediante la función `solve` de `NumPy` si lo aplicamos a las matrices 
$$A=\left(\begin{array}{rrr}
1 & 0 & 0 \\
2 & 2 & 0 \\
3 & 1 & 5 \end{array} \right) \qquad b=\left(2,5,1 \right) $$.

```python
import numpy as np

def sust_progresiva(A,b):   #función modificada del documento práctica 1
    """
    Esta función resuelve el sistema Ax=b mediante sustitución progresiva. 
    A y b deben ser arrays de numpy. 
    A debe ser una matriz triangular inferios.
    """
    # A=A.astype(float)     # Con esto se cambian los elementos a tipo flotante pero no es necesario
    # b=b.astype(float) 
    x=np.zeros(np.size(b))         # almacenaremos aquí la solución del sistema
                       
    # ahora se resuelve el sistema por sustitución progresiva
    x[0]=b[0]/A[0,0]
    for i in range(np.size(b)):     
        x[i]=(b[i]-np.dot(A[i][:i],x[:i]))/A[i,i]
                
    return x


A1=np.array([[1,0,0],[2,2,0],[3,1,5]])
b1=np.array([2,5,1])

x1=sust_progresiva(A1,b1) # llamo a la función definida previamente
x2=np.linalg.solve(A1,b1) # pruebo la función solve

print (f'Solución sustitución regresiva:  {x1}')
print (f'Solución función solve de NumPy: {x2}') #verifica si está hecho correctamente

if(x1.all()==x2.all()):
    print ('La función SI se corresponde con el resultado')
else:
    print ('La función NO se corresponde con el resultado')
```

**Ejercicio 4:** Utilizando la descomposición LU de la práctica 2 y las sustituciones progresiva y regresiva de la práctica 1, resuelve
de nuevo el sistema del ejercicio 2.

```python
DEBUG=0

def sust_regresiva(A,b):
    """
    Esta función resuelve el sistema Ax=b mediante sustitución regresiva. 
    A y b deben ser arrays de numpy. 
    A debe ser una matriz triangular superior.
    """
    # A=A.astype(float)     # Con esto se cambian los elementos a tipo flotante pero no es necesario
    # b=b.astype(float) 
    n=np.size(b)
    x=np.zeros(n)         # almacenaremos aquí la solución del sistema
                       
    # ahora se resuelve el sistema por sustitución regresiva
    x[n-1]=b[n-1]/A[n-1,n-1]
    for i in range(n-2,-1,-1):     
        x[i]=(b[i]-np.dot(A[i][i+1:],x[i+1:]))/A[i,i]
                
    return x

def sust_progresiva(A,b):   #función del ejercicio 1
    """
    Esta función resuelve el sistema Ax=b mediante sustitución progresiva. 
    A y b deben ser arrays de numpy. 
    A debe ser una matriz triangular inferios.
    """
    # A=A.astype(float)     # Con esto se cambian los elementos a tipo flotante pero no es necesario
    # b=b.astype(float) 
    x=np.zeros(np.size(b))         # almacenaremos aquí la solución del sistema
                       
    # ahora se resuelve el sistema por sustitución progresiva
    x[0]=b[0]/A[0,0]
    for i in range(np.size(b)):     
        x[i]=(b[i]-np.dot(A[i][:i],x[:i]))/A[i,i]
                
    return x

def LU_desc(A):             #función del documento de prácticas
    """
    Esta función calcula la descomposición LU de Doolittle de una matriz A sin intercambio de filas.
    A debe ser un array de Numpy.
    """
    n=np.shape(A)[1]
    U=np.zeros([n,n])
    L=np.eye(n)
    
    for k in range(n-1):
        U[k,k]=A[k,k]-np.dot(L[k,:k],U[:k,k])
        for j in range(k+1,n):
            U[k,j]=A[k,j]-np.dot(L[k,:k],U[:k,j])
        for j in range(k+1):
            L[k+1,j]=(A[k+1,j]-np.dot(L[k+1,:j],U[:j,j]))/U[j,j]
            
    U[n-1,n-1]=A[n-1,n-1]-np.dot(L[n-1,:n-1],U[:n-1,n-1])
    
    return L, U


L4,U4 = LU_desc(A2)

if(DEBUG==1):
    print('Matriz A:')
    print(A2)
    print()
    print('Descomposicion LU:')
    print('L:')
    print(L4)
    print()
    print('U:')
    print(U4)

y4=sust_progresiva(L4,b2) # Resuelvo el sistema triangular inferior
x4=sust_regresiva(U4,y4)  # Resuelvo el sistema triangular superior

print ('Solucion con funcion linalg:')
print(np.linalg.solve(A2,b2))
print()
print('Solucion con LU:')
print(x4)

```

```python
DEBUG=0

def sust_regresiva(A,b):
    """
    Esta función resuelve el sistema Ax=b mediante sustitución regresiva. 
    A y b deben ser arrays de numpy. 
    A debe ser una matriz triangular superior.
    """
    # A=A.astype(float)     # Con esto se cambian los elementos a tipo flotante pero no es necesario
    # b=b.astype(float) 
    n=np.size(b)
    x=np.zeros(n)         # almacenaremos aquí la solución del sistema
                       
    # ahora se resuelve el sistema por sustitución regresiva
    x[n-1]=b[n-1]/A[n-1,n-1]
    for i in range(n-2,-1,-1):     
        x[i]=(b[i]-np.dot(A[i][i+1:],x[i+1:]))/A[i,i]
                
    return x

def sust_progresiva(A,b):   #función del ejercicio 1
    """
    Esta función resuelve el sistema Ax=b mediante sustitución progresiva. 
    A y b deben ser arrays de numpy. 
    A debe ser una matriz triangular inferior.
    """
    # A=A.astype(float)     # Con esto se cambian los elementos a tipo flotante pero no es necesario
    # b=b.astype(float) 
    x=np.zeros(np.size(b))         # almacenaremos aquí la solución del sistema
                       
    # ahora se resuelve el sistema por sustitución progresiva
    x[0]=b[0]/A[0,0]
    for i in range(np.size(b)):     
        x[i]=(b[i]-np.dot(A[i][:i],x[:i]))/A[i,i]
                
    return x

def Cholesky (A):
    """
    Esta función realiza la factorizaciñon de Cholesky que consiste en 
    A = L1 * L2
    donde A es simétrica, L1 es triangular inferior y L2 es L1 traspuesta
    """
    n=np.shape(A)[0] #toma el valor del número de filas de A
    L=np.zeros([n,n])
    
    for k in range (n):
        
        sumatory1=0
        
        for s in range (k):
            sumatory1+=((L[k][s])**2)
        
        L[k][k]=((A[k][k]-sumatory1)**0.5)
        
        for j in range (k+1,n):
            sumatory2=0
            
            for s in range (k):
                sumatory2+=(L[j][s]*L[k][s])
            
            L[j][k]=(A[j][k]-sumatory2)/L[k][k]
            
    Lt=np.transpose(L)
            
    return L,Lt

A51=np.array([[2,4,1],[4,10,1],[1,1,5]])
A52=np.array([[2,4,1],[4,0,1],[1,1,5]])

b5=np.array([2,-5,1])

L51 , Lt51=Cholesky(A51)
L52 , Lt52=Cholesky(A52)
L53 , Lt53=Cholesky(A2)

if(DEBUG==1):
    print('Matriz A:')
    print(A51)
    print()
    print('Matriz L:')
    print(L51)
    print('Matriz Lt')
    print(Lt51)
    print()
    print('L*Lt=')
    print(np.dot(L51,Lt51))
    print()
    print('----------------------')
    
    print('Matriz A:')
    print(A2)
    print()
    print('Matriz L:')
    print(L53)
    print('Matriz Lt')
    print(Lt53)
    print()
    print('L*Lt=')
    print(np.dot(L53,Lt53))
    print()
    print('----------------------')

y51=sust_progresiva(L51,b5) # Resuelvo el sistema triangular inferior
x51=sust_regresiva(Lt51,y51)  # Resuelvo el sistema triangular superior

y52=sust_progresiva(L52,b5) # Resuelvo el sistema triangular inferior
x52=sust_regresiva(Lt52,y52)  # Resuelvo el sistema triangular superior

y53=sust_progresiva(L53,b2) # Resuelvo el sistema triangular inferior
x53=sust_regresiva(Lt53,y53)  # Resuelvo el sistema triangular superior

print('SISTEMA 1:')
print ('Solucion con funcion linalg:')
print(np.linalg.solve(A51,b5))
print()
print('Solucion con Cholesky:')
print(x51)
print('---------------')

print('SISTEMA 2:')
print ('Solucion con funcion linalg:')
print(np.linalg.solve(A52,b5))
print()
print('Solucion con Cholesky:')
print(x52) #no se puede resolver el sistema por Cholesky ya que A no es definida positiva

# El menor principal de orden 2 es:

# | 2 4 |
# | 4 0 |  = 2*0 - 4*4 = 0 - 16 = -16 < 0  ==> A no es definida positiva

print('---------------')

print('SISTEMA 3:')
print ('Solucion con funcion linalg:')
print(np.linalg.solve(A2,b2))
print()
print('Solucion con Cholesky:')
print(x53) #no se puede resolver el sistema por Cholesky ya que A no es definida positiva

# Para probarlo veamos 

for i in range (np.shape(A2)[1]):
    print(f'Menor principal de orden {i+1}: {np.linalg.det(A2[:i+1,:i+1])}')

print('---------------')

```

**Ejercicio 6:** Modifica el código para la resolución de un sistema mediante el método de Jacobi (práctica 3) para resolver de
forma aproximada un sistema de ecuaciones lineales utilizando el método de Gauss-Seidel.

```python
def Gauss_Seidel(A, b, maxiter=1000, tol=10**(-10)):
    """
    Esta función resuelve el sistema Ax=b de forma aproximada mediante el método iterativo de Gauss-Seidel empezando las iteraciones
    en el vector nulo.
    Las matrices A y b deben ser arrays de NumPy.
    El número máximo de iteraciones se puede introducir por el usuario y es 1000 por defecto.
    La tolerancia también se puede introducir por el usuario y es 10^(-10) por defecto.
    El método se interrumpe cuando se alcanza el número máximo de iteraciones o cuando la tolerancia es menor que la indicada.
    
    """
    
    n=np.shape(A)[1] # se calcula la dimensión de A
    d=np.diag(A)     # obtenemos un array con los elementos de la diagonal de la matriz A
    x=np.zeros(n)
    y=np.zeros(n)
          
    if np.prod(d)==0:
        print('Hay elementos nulos en la diagonal de la matriz')
    else:
        k=1
        er=1
        nx=1
        while (er>tol*nx and k<maxiter):
            nx=np.linalg.norm(x,1)        
            for i in range(n):
                y[i] = (b[i] - np.dot(A[i,:i],y[:i]) - np.dot(A[i,i+1:],x[i+1:]) )/A[i,i]  
            er=np.linalg.norm(y-x,1)   # calculamos la norma de la diferencia entre y y el resultado de la iteración anterior        
            x=np.copy(y)
           # print('Iteración k=', k, '  x^(k)= ', x ) # descomentar si se quieren ver las iteraciones.
            k+=1
        print(f'Iteraciones: {k-1}')
    return x
    
A6=np.array([[10,4,1],[4,10,1],[1,1,5]])
b6=np.array([15,15,7])

print(Gauss_Seidel(A6,b6,13,10**(-5)))
```
