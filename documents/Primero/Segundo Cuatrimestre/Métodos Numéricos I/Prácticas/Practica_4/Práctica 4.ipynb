```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-1,1,50)

def f(x):
    return (x**2+1)

y=f(x)

#plt.plot(x,y)
#plt.plot(x,y,'r3')
#plt.show()

xf=np.linspace(-1,1,18)
yf=f(xf)

plt.plot(xf,yf,'g',label='datos')
plt.plot(xf,yf,'r.',label='$x^2+1$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color='lightblue')
plt.title('Mi primer gráfico',color='purple', loc='center')
plt.legend()
plt.show()
```

**Ejercicio:** Dibuja las funciones: $sin(\pi x)$ y $cos(\pi x)$ en el intervalo $[-1,1]$


```python

x1=np.linspace(-1,1,100)

y1=np.sin(np.pi*x1)
y2=np.cos(np.pi*x1)

plt.plot(x1,y1,'g',label='datos')
plt.plot(x1,y1,'r.',label='$sin(\pi x)$')

plt.plot(x1,y2,'b',label='datos')
plt.plot(x1,y2,'g.',label='$sin(\pi x)$')

plt.xlabel('x')
plt.ylabel('y')
plt.grid(color='lightblue')
plt.title('Ejercicio 1',color='purple', loc='center')
plt.legend()
plt.figure(figsize=[15, 6])
plt.show()
```
